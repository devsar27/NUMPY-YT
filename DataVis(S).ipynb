```python
import seaborn as sns
```

```python
###Basic Plotting with Seaborn
tips = sns.load_dataset("tips")
tips
```

```python
##Scatter Plot
import matplotlib.pyplot as plt
sns.scatterplot(x = "total_bill", y = "tip", data = tips)
plt.title("Scatter Plot of Total Bill vs Tip")
plt.show()
```

```python
##Line Plot
sns.lineplot(x = "size", y = "total_bill", data = tips)
plt.title("Line Plot of Size vs Total Bill")
plt.show()
```

```python
##Categorial Plot
##Bar plot
sns.barplot(x="day", y= "total_bill",data = tips)
plt.title("Bar Plot of Day vs Total Bill")
plt.show()
```

```python
##Box plot 
sns.boxplot(x="day", y="total_bill", data=tips)
plt.title("Box Plot of Day vs Total Bill")
plt.show()
```

```python
##Violin Plot
sns.violinplot(x="day", y="total_bill", data=tips)
plt.title("Violin Plot of Day vs Total Bill")
plt.show()
```

```python
##Histogram
sns.histplot(tips["total_bill"], bins=10, kde=True)   
##KDE = Kernel Density Estimate
##It is a smooth curve that estimates the probability density of a continuous variable.
plt.title("Histogram of Total Bill")
plt.show()
```

```python
##KdE Plot
sns.kdeplot(tips["total_bill"], fill=True)
plt.title("KDE Plot of Total Bill")
plt.show()
```

```python
##Pair Plot
sns.pairplot(tips)
```

```python
##Heat Map 
corr = tips[["total_bill",'tip', 'size']].corr()
corr 
```

```python
sns.heatmap(corr, annot=True, cmap="YlGnBu")
plt.title("Heatmap of Correlation Matrix")
plt.show()
```
