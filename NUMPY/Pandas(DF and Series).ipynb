```python
import pandas as pd 
data = [1,2,3,4,5]
series = pd.Series(data)
print("Pandas Series:\n", series)
```

```python
## Create a Series from a dictionary
data = {'a': 10, 'b': 20, 'c': 30}
series_dict = pd.Series(data)
print("\nPandas Series from dictionary:\n", series_dict)
```

```python
data = [1,2,3]
index = ['x','y','z']
series_indexed = pd.Series(data, index= index)
print("\nPandas Series with custom index:\n", series_indexed)
```

```python
## DataFrame
## create a DataFrame from a dictionary of lists

data ={
    'Name': ['Alice', 'Bob', 'Charlie'],
    'Age': [25, 30, 35],
    'City': ['New York', 'Los Angeles', 'Chicago']
}
df = pd.DataFrame(data)
print("\nPandas DataFrame:\n", df)
print(type(df))
```

```python
##Create a DataFrame from a list of dictionaries
data = [
    {'Name': 'David', 'Age': 28, 'City': 'Miami'},
    {'Name': 'Eva', 'Age': 22, 'City': 'Seattle'},
    {'Name': 'Frank', 'Age': 33, 'City': 'Boston'}
]
df=pd.DataFrame(data)
print("\nPandas DataFrame from list of dictionaries:\n", df)
print(type(df))
```
