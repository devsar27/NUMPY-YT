```python
import numpy as np 
```

```python
arr = np.random.randint(1,21, size=(5,5))
print("Original Array:\n", arr)

arr[:,2] = 1
print("Modified Array:\n", arr)
```

```python
arr = np.arange(1, 17).reshape((4, 4))
print("Original Array:\n", arr)

np.fill_diagonal(arr, 0)
print("Modified Array:\n", arr)
```
